####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
import seaborn as sns
import matplotlib.pyplot as plt

ax = plt.subplots()
plt.close('all')
plt.rc('figure', figsize=(11, 10))
fig, axes = plt.subplots(3, 2)

sns.countplot(ax = axes[0, 0],x = 'APGAR1', data=dados)
sns.countplot(ax = axes[0, 1],x = 'APGAR5', data=dados)
sns.countplot(ax = axes[1, 0],x = 'SEXO', data=dados)
sns.countplot(ax = axes[1, 1],x = 'ESTCIVMAE', data=dados)
sns.countplot(ax = axes[2, 1],x = 'GESTACAO', data=dados)
plt.xticks(rotation=45)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sns.barplot(x="APGAR5", y='CONSULTAS', data=dados)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS
